```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read, write
from numpy.fft import fft, ifft
from IPython.display import Audio

FRAME_SIZE = 1024

################## YOUR CODE HERE ######################
def ece420ProcessFrame(frame, Fs, N):
#     N = FRAME_SIZE
    # Check voiced/unvoiced
    isVoiced = 0
    E = 0
    
    for i in range(N):
        x = abs(frame[i])
        x = x * x
        E += x
    
    if E > 1800000000:
        isVoiced = 1
        
    if not isVoiced:
        return -1
    
    # Compute autocorrelation
    fft_1 = np.fft.fft(frame)
    autocor1 = np.fft.ifft(fft_1 * np.conj(fft_1))
    end = int(FRAME_SIZE / 2)
    # end = autocor1.shape[0]
    peaks = np.zeros(end)
    
    # Find average
    avg = 0
    for i in range(20, end):
        # print(i)
        avg += autocor1[i]
    avg = avg / float(end - 20)

    # Find end of zero neighborhood
    start = 0
    while autocor1[start] >= avg:
        start +=1
    
    peak_inds = []
    curr_ind = 0
    curr_val = 0
    
    # Multiple peak detection
    for i in range(start, end):
        val = autocor1[i]
        if val > avg:
            if curr_val == 0 or val > curr_val:
                curr_ind = i
                curr_val = val
        elif (val < avg and curr_val != 0):
            peak_inds.append((curr_ind, curr_val))
            curr_ind = 0
            curr_val = 0
    
    if curr_ind != 0:
        peak_inds.append((curr_ind, curr_val))
    
    peak_inds = np.array(peak_inds)
    A = peak_inds.shape[0]
    sorted = peak_inds[peak_inds[:, 1].argsort()]
    top = sorted[(A-3):]

    freq = []
    # print(top.shape[0])
    for i in range(top.shape[0]):
        freq.append(Fs / float(top[i][0]))

    # if len(freq) >= 3:
    #     return freq[2]
    # else:
    # print(freq[0])
    return freq[2]


################# GIVEN CODE BELOW #####################

Fs, data = read('majorthird.wav')

# Audio(data,rate=Fs)

numFrames = int(len(data) / FRAME_SIZE)
# print(numFrames)

frequencies = np.zeros(numFrames)
# autocor1 = np.zeros(FRAME_SIZE)
# for i in range(numFrames):
#     frame = data[i * FRAME_SIZE : (i + 1) * FRAME_SIZE]
#     frequencies[i] = ece420ProcessFrame(frame.astype(float), Fs, FRAME_SIZE)
#     if i == 210:
#         fft_1 = np.fft.fft(frame)
#         autocor1 = np.fft.ifft(fft_1 * np.conj(fft_1))
#         end = int(FRAME_SIZE / 2)
#         # end = autocor1.shape[0]
#         peaks = np.zeros(end)
        
#         # Find average
#         avg = 0
#         for i in range(20, end):
#             # print(i)
#             avg += autocor1[i]
#         avg = avg / float(end - 20)
    
#         # Find end of zero neighborhood
#         start = 0
#         while autocor1[start] >= avg:
#             start +=1
        
#         peak_inds = []
#         curr_ind = 0
#         curr_val = 0
        
#         # Multiple peak detection
#         for i in range(start, end):
#             val = autocor1[i]
#             if val > avg:
#                 if curr_val == 0 or val > curr_val:
#                     curr_ind = i
#                     curr_val = val
#             elif (val < avg and curr_val != 0):
#                 peak_inds.append((curr_ind, curr_val))
#                 curr_ind = 0
#                 curr_val = 0
        
#         if curr_ind != 0:
#             peak_inds.append((curr_ind, curr_val))
        
#         peak_inds = np.array(peak_inds)
#         A = peak_inds.shape[0]
#         sorted = peak_inds[peak_inds[:, 1].argsort()]
#         top = sorted[(A-3):]
#         freq = []
#         for i in range(top.shape[0]):
#             freq.append(Fs / float(top[i][0]))

# # PLOTTING
# print(np.average(frequencies))
# plt.figure(figsize=(8,25))
# plt.subplot(5,1,1)
# plt.plot(np.abs(autocor1))
# plt.scatter(top[:,0], top[:,1], color = 'red')
# plt.axis('tight')
# plt.xlabel('Frame idx')
# plt.ylabel('Hz')
# plt.title('Detected Frequencies in Hz')


# ORIGINAL PLOTTING
for i in range(numFrames):
    frame = data[i * FRAME_SIZE : (i + 1) * FRAME_SIZE]
    frequencies[i] = ece420ProcessFrame(frame.astype(float), Fs, FRAME_SIZE)

# PLOTTING
plt.figure(figsize=(8,25))
plt.subplot(5,1,1)
plt.plot(frequencies)
plt.axis('tight')
plt.xlabel('Frame idx')
plt.ylabel('Hz')
plt.title('Detected Frequencies in Hz')

plt.show()
```
